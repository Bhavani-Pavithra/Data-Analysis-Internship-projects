```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.dates import DateFormatter
```

```python
df = pd.read_csv("SampleSuperstore.csv")  # or change filename to your actual one
df.head()
```

```python
plt.figure(figsize=(8,5))
sns.barplot(x='Category', y='Sales', data=df, estimator=sum, ci=None, palette='pastel')
plt.title("Total Sales by Category")
plt.ylabel("Sales")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(12,5))
sns.barplot(x='Sub-Category', y='Sales', data=df, estimator=sum, ci=None, palette='muted')
plt.title("Total Sales by Sub-Category")
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(data=df, x='Sales', y='Profit', hue='Category', style='Region')
plt.title("Profit vs Sales")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.barplot(x='Region', y='Sales', data=df, estimator=sum, ci=None, palette='coolwarm')
plt.title("Total Sales by Region")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='Category', y='Discount', data=df, palette='Set3')
plt.title("Discount Distribution by Category")
plt.grid(True)
plt.show()

plt.figure(figsize=(8,5))
sns.scatterplot(x='Discount', y='Quantity', data=df, hue='Category')
plt.title("Discount vs Quantity")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.heatmap(df[['Sales', 'Profit', 'Discount', 'Quantity']].corr(), annot=True, cmap='YlGnBu')
plt.title("Correlation Matrix")
plt.show()
```
